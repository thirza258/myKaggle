```python
df['gaia_id'].fillna(0, inplace=True)
df = df[df['gaia_id'] != 0]
df.shape
```

```python
df['releasedate'] = pd.to_datetime(df['releasedate'])
df
```

```python
df['Month_Year'] = df['releasedate'].dt.to_period('M')
count_data= df['Month_Year'].groupby(df['Month_Year']).count().reset_index(name='count')
count_data = count_data[count_data['Month_Year'] < '2023-08']
# df['Month_Year'].describe() 
count_data.plot(x='Month_Year', y='count', figsize=(15,5), grid=True, title='Number of Exoplanets Discovered by Month/Year')
```
